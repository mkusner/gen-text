```python
from __future__ import division

%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

sin = np.sin
exp = np.exp
```

```python
domain = np.linspace(-10,10,1000)
actual = lambda x: 1./3 + x + sin(x*x) 

res_grammar = (lambda x: x/1+sin(3)+sin(x*x),
               lambda x: 1/2+(x)+sin(x*x),
               lambda x: x/x+(x)+sin(x*x))
res_character = (lambda x: x*1+sin(3)+sin(3/1),
                 lambda x: x*1+sin(1)+sin(2*3),
                 lambda x: x+1+sin(3)+sin(3+1))
```

```python

#x+1+exp(1)+sin(2*2)
plt.figure(figsize=(6.5,4))
plt.plot(domain, actual(domain), 'k-')

for eq in res_grammar[:1]:
    plt.plot(domain, eq(domain), '-', color='#3333CC')
    
for eq in res_character[:1]:
    plt.plot(domain, eq(domain), '-', color='#CC3333')

plt.legend(['Actual function', 'Best GVAE', 'Best CVAE'], loc='best')
plt.ylim(-12,12)
#plt.plot(domain, earlier(domain))
# plt.plot(domain, found(domain))
# plt.plot(domain, char(domain))
# plt.legend(['actual', 'eq1', 'eq2', 'char'], loc='best')
plt.savefig("learned-functions.pdf")
```
